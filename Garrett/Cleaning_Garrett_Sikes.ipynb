```python
import numpy as np
import pandas as pd 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../../accepted_2007_to_2018Q4.csv.gz')
df.drop(columns='member_id', inplace=True)

df = df[df['issue_d'].notna()]

df.shape
```

## Clean Missing Variables

```python
df['annual_inc_joint'].fillna(df['annual_inc'], inplace=True)
df = df[df['annual_inc_joint'].notna()]
df['dti_joint'].fillna(df['dti'], inplace=True)
df = df[df['dti_joint'].notna()]
```

```python
mode_ = ['emp_length', 'inq_fi', 'inq_last_12m', 'open_acc', 'pub_rec']

for col in mode_:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
zero_ = ['pub_rec_bankruptcies', 'total_acc', 'revol_util', 'delinq_amnt', 'delinq_2yrs',\
         'mths_since_last_delinq', 'mths_since_last_major_derog']

for col in zero_:
    df[col].fillna(0, inplace=True)
```

```python
df_raw = df[['issue_d', 'grade', 'loan_status', 'annual_inc_joint', 
        'dti_joint', 'emp_length', 'fico_range_high',
       'fico_range_low', 'funded_amnt', 'home_ownership', 'inq_fi',
       'inq_last_12m', 'installment', 'int_rate', 'loan_amnt', 'open_acc',
       'pub_rec', 'pub_rec_bankruptcies', 'total_acc',
       'revol_bal', 'revol_util', 'delinq_amnt', 'delinq_2yrs',
       'mths_since_last_delinq', 'mths_since_last_major_derog', 'purpose']].copy()

```

```python
df_raw['issue_d'] = pd.to_datetime(df_raw['issue_d'], format='%b-%Y')
df_raw['issue_month'] = df_raw['issue_d'].dt.month
df_raw['issue_year'] = df_raw['issue_d'].dt.year

df_raw.drop(columns='issue_d', inplace=True)
```
